```python
import numpy as np
import torch
import torchvision
```

```python
# Creating a simple feedforward neuran network
class FNN(torch.nn.Module):
    def __init__(self):
        torch.nn.Module.__init__(self)
        
        self.f1 = torch.nn.Linear(2,5)
        self.relu1 = torch.nn.ReLU()
        self.f2 = torch.nn.Linear(5,2)
        self.relu2 = torch.nn.ReLU()
        
    def forward(self,x):
        o = self.f1(x)
        o = self.relu1(o)
        o = self.f2(o)
        o = self.relu2(o)
        return o
    
    def train(self,train_dl,epochs=20):
        opt = torch.optim.SGD(self.parameters(),lr=1)
        crit = torch.nn.MSELoss()
        
        for epoch in range(epochs):
            for i,(x,y) in enumerate(train_dl):
                opt.zero_grad()
                o = self.forward(x)
                loss = crit(o,y)
                print(i,loss)
                loss.backward()
                opt.step()
                
            
```

```python
x_train = torch.Tensor([[0,0],[0,1],[1,0],[1,1]]).float()
y_train = torch.Tensor([0,1,1,0]).long()
```

```python
ds = torch.utils.data.TensorDataset(x_train,y_train)
dl = torch.utils.data.DataLoader(ds,batch_size=4)
```

```python
model = FNN()
model.forward(x_train)
```

```python
model.train(dl,epochs=1000)
```
